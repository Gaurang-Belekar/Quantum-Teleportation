```python
from qiskit import *
```

```python
circuit = QuantumCircuit(3,3) # Quantum Circuit madeup of 3 qubit and 3 classical bit
```

```python
%matplotlib inline
circuit.draw(output = 'mpl')
```

```python
circuit.x(0)
```

```python
circuit.barrier()
circuit.draw(output = 'mpl')
```

```python
#Now for transfering data from q0 to q2, we will first entangle q1 and q2
```

```python
circuit.h(1)
circuit.cx(1,2)
```

```python
circuit.draw(output = 'mpl')
```

```python
circuit.barrier()
```

```python
#We have entangled q1 and q2. Now with the help of q1,we will teleport q0 information to q2.
```

```python
#So first we will measure the entagled q1 state with information of q0
```

```python
circuit.cx(0,1)
circuit.h(0)
```

```python
circuit.draw(output='mpl')
```

```python
#Now after doing bell measurement, we will now measure the info of q0 and q1 ,and store it in classical bits
```

```python
circuit.measure([0,1], [0,1])
```

```python
circuit.draw(output ='mpl')
```

```python
#Now we will pass the information of q0 to q1 by measuring entangled q2, with q0
```

```python
circuit.barrier()
circuit.cx(1,2)
circuit.cz(0,2)
```

```python
circuit.draw(output='mpl')
```

```python
#Now checking the outcome by measuring q2
```

```python
circuit.draw(output ='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit,backend = simulator, shots = 1024).result()
counts = result.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```

```python
# Here we see that as our input information was ket(1) and now after doing the teleportation protocol, we see that the information is transfered completely to q2 = ket(1) for any measurement of q0 and q1
```
